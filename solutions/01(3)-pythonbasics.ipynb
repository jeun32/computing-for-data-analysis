Write a function that returns the students names from "top to bottom".

    ### Solution - Exercise 3  
    def build_grade_dicts(grades: list) -> dict:
        ### BEGIN SOLUTION
        # assignments = get_assignments(grades)
        assignments = grades[0][1:]
        return {L[0]: dict(zip(assignments, [int(g) for g in L[1:]])) for L in grades[1:]}

    ### Solution - Exercise 4  
    def build_avg_by_student(grades: list) -> dict:
        ### BEGIN SOLUTION
        # grade_lists = build_grade_lists(grades)
        grade_lists={L[0]:[int(g) for g in L[1:]] for L in grades[1:]}
        from statistics import mean
        return {n: round(mean(G),3) for n, G in grade_lists.items()}
        ### END SOLUTION

    def build_avg_by_asn(grades: list) -> dict:
